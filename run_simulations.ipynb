# Trolley simulation

```python
dt = 0.01
time = np.arange(0, 200, dt)
target = np.ones_like(time) * 10
disturbance = np.zeros_like(time)


# Create a PID controller
pid = PID(KP=1, KI=0.5, KD=1)

# Create a system
system = Trolley(mass=2, friction=0.2, dt=dt)

# Create a simulation environment
sim = Simulation(time, target, disturbance, dt)
```

Run the simulation

```python
# Run the simulation
sim.run(system, pid)
sim.plot()
```
